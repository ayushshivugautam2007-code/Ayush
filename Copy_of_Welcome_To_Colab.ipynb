```
#while loop
i = 0
while i<5:
  print(i)
  i+=1
```

```
#loop control statement
#break-- this statement break the loop
i = 0
while i<5:
  i+=1
  if i==3:
    break
  else:
      print(i)
```

```
#continue-- this statement sip the current iteration
i = 0
while i<5:
  i+=1
  if i==3:
     continue
  else:
      print(i)
```

```
#pass-- pass the line
i = 0
while i<5:
  i+=1
  if i==3:
    pass
    print(i)
  else:
      print(i)
```

```
#non keyword variable length argument
def test(*args):
  print(args)
  print(len(args))
  print(type(args))
```

```
def sum_number(*args):
  sum = 0
  for num in args:
    sum+=num
  return sum
```

```
sum_number(1,2,3,4,5,6)
```

```
#keyword variable length argument
def test(**x):
  print(x)
  print(len(x))
  print(type(x))
```

```
test(a=10,b=20,c=30)
```

```
def pass_student_list(**x):
  pass_student_list = []
  for name, marks in x.items():
    if marks>35:
      pass_student_list.append(name)
      return pass_student_list
```
